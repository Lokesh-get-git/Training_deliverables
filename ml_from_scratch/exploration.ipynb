# Linear Regression Demo

```python
import pandas as pd
import matplotlib.pyplot as plt
import linear_regression as lr
```

```python
df=pd.read_csv("synthetic_data/linear_regression_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df.score,df.study_time)
plt.show()
```

```python
lrmodel=lr.linear_model()
m,b=lrmodel.build_model(df.study_time,df.score)
```

```python
lrmodel.predict(56.71117368334277)
```

```python
plt.scatter(df.study_time,df.score)
plt.plot(list(range(20,80)),[ m*x+b for x in range(20,80)],color="red")
plt.show()
```
